```python
import numpy as np
import cv2
np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
test_images = np.array([cv2.imread("C:\\Users\\user\\Desktop\\test\\testimg1.jpg"), cv2.imread("C:\\Users\\user\\Desktop\\test\\testimg2.jpg"), ...])
ground_truth_labels = np.array([1, 0, ...])
```

```python
cascade_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
import glob
testP=["C:\\Users\\user\\Desktop\\test\\testimg4.jpg","C:\\Users\\user\\Desktop\\test\\testimg2.jpg","C:\\Users\\user\\Desktop\\test\\testimg3.jpg","C:\\Users\\user\\Desktop\\test\\testimg3.jpg"]
dir_path = r'C:\\Users\\user\\Desktop\\test\\data\**\*.*'
for file in glob.glob(dir_path, recursive=True):
    testP.append(file)
```

```python
print(testP)
predicted_labels = []

for i in testP:
    img = cv2.imread(i)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    faces = cascade_classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
    if len(faces) > 0:
        predicted_labels.append(1)
    else:
        predicted_labels.append(0)
```

```python

print(predicted_labels)
print(ground_truth_labels)
ground_truth_labels=ground_truth_labels[:1]
x =(predicted_labels - ground_truth_labels)
mse = np.mean(x**2)

print("Mean Squared Error:", mse)
```

```python
y_pred=tuple(predicted_labels)

print(y_pred)
```

```python
y_true = list(y_pred)


y_true[0]=1
y_true[232]=1
y_true[243]=1
# 46 false  

pVal = list(y_pred)

print(y_true)
print(pVal)

```

```python
count=0
for i in range(0,len(y_true)):
    if y_true[i]!=pVal[i]:
        count+=1

print(count)
print(len(y_true))

print((len(y_true)-count)/len(y_true))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score


precision = precision_score(y_true, pVal)
recall = recall_score(y_true, pVal)

f1 = f1_score(y_true,pVal)

print("Precision: {:.3f}".format(precision))
print("Recall: {:.3f}".format(recall))
print("F1 score: {:.3f}".format(f1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example true labels and predicted probabilities
print(y_true,pVal)

# Calculate the log loss for each example
log_losses = []
for i in range(len(y_true)):
    y = pVal[i]
    if y == 0:
        y = 1e-15
    elif y == 1:
        y = 1 - 1e-15
    log_loss = - (y_true[i] * np.log(y) + (1 - y_true[i]) * np.log(1 - y))
    log_losses.append(log_loss)

# Calculate the average log loss
average_log_loss = np.mean(log_losses)

# Create the log loss graph
plt.plot(range(len(y_true)), log_losses, marker='o')

# Set the axis labels and title
plt.xlabel('Example')
plt.ylabel('Log Loss')
plt.title('Log Loss vs. Example')

# Set the axis ranges
plt.ylim([0, max(log_losses) * 1.1])

# Show the plot
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Example true labels and predicted probabilities


# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_true, pVal)

# Calculate the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)

# Create the ROC curve
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")

# Show the plot
```
